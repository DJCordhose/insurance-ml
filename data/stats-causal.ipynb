```python
# proba of well dressed
p_z = 0.5

# well dresses
z = np.random.binomial(n=1, p=p_z, size=500)
z.mean(), z.std()
```

```python
z
```

```python
# 
p_x_z = [0.9, 0.1]
p_x = np.choose(z, p_x_z)

# probas of 
p_x
```

```python
x = np.random.binomial(n=1, p=p_x, size=500)
x.mean(), x.std()
```

```python
x
```

```python
index = x+2*z
index
```

```python
p_y_xz = [0.2, 0.4, 0.6, 0.8]
p_y = np.choose(index, p_y_xz)
p_y
```

```python
y = np.random.binomial(n=1, p=p_y, size=500)
y
```

```python
# x: properly dressed
# y: productive

generate_dataset_0 = pd.DataFrame({"x":x, "y":y, "z":z})
generate_dataset_0.describe()
```

```python
ds = generate_dataset_0
ds
```

```python
ds_not_proper = ds[ds.x == 0]
ds_not_proper.count()
```

```python
ds_not_proper.y.sum(), ds_not_proper.y.mean()
```

```python
ds_proper = ds[ds.x == 1]
ds_proper.count()
```

```python
ds_proper.y.sum(), ds_proper.y.mean()
```

```python
def estimate_uplift(ds):
    base = ds[ds.x == 0]
    variant = ds[ds.x == 1]
    delta = variant.y.mean() - base.y.mean()
    delta_err = 1.96 * np.sqrt(
        variant.y.var() / variant.shape[0] + 
        base.y.var() / base.shape[0])
    return {"estimated_effect": delta, "standard_error": delta_err}
estimate_uplift(generate_dataset_0)
```
